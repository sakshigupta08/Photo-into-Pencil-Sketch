# Photo into Sketch

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('d.jpg')
```

```python
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img,cmap='gray')
```

```python
invert=255-gray_img
```

```python
plt.imshow(invert,cmap='gray')
```

```python
blurr=cv2.GaussianBlur(invert,(21,21),0)
invert_blur=255-blurr
```

```python
pencil_sketch=cv2.divide(gray_img,invert_blur,scale=256.0)
```

```python
plt.imshow(pencil_sketch,cmap='gray')
```
